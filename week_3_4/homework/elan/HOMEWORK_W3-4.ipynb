````kotlin
val products = listOf(
    mapOf("name" to "Keyboard", "price" to 50000),
    mapOf("name" to "Egg", "price" to 5000),
    mapOf("name" to "PineApple", "price" to 15000),
    mapOf("name" to "Pizza", "price" to 15000),
    mapOf("name" to "Bread", "price" to 3000),
    mapOf("name" to "Onion", "price" to 1200),
    mapOf("name" to "Milk", "price" to 3000),
)
val cart = mutableListOf<Map<String, Any>>()

fun showAllProducts(): Unit {
    products.mapIndexed { i, entry ->
        "${i + 1} . ${entry["name"]}: ${entry["price"]}"
    }.forEach { println(it) }
    println("")
}

fun addCart(no: Int): Unit {
    val entry = products[no]
    cart.add(entry).also {
        println("상품이 장바구니에 추가 되었습니다. ${entry["name"]}: ${entry["price"]}")
    }

}

fun printProduct(entry: Map<String, Any>): Unit = println("${entry["name"]}: ${entry["price"]}")

fun showResult(): Unit {
    println("\n최종 장바구니 목록")
    cart.forEach { printProduct(it) }
    cart.sumOf({ it["price"].toString().toInt() })
        .let {
            print("총 금액 $it")
            if (it > 100_000) {
                println(" - 할인 적용 대상입니다!")
            }
        }
    cart.filter {
        it["price"].toString().toInt() < 10_000
    }.let {
        println("\n10,000원 이하 상품 목록:")
        it.forEach { item -> printProduct(item) }
    }
}

showAllProducts()
while(true) {
    val input = readlnOrNull()?.trim()
    if (input == null || input.lowercase() == "n") {
        break
    }
    addCart(input.toInt() - 1)
}
showResult()
````

- 대량의 데이터 다운로드 시뮬레이션

요구사항
- 1~100 번의 번호를 가진 파일을 다운로드 한다고 가정한다.
- 한번에 5개의 파일을 다운로드 할 수 있다고 가정한다. (chunked 나 semaphore 사용)
- 다운로드는 500 ms 가 걸린다고 가정 한다 (deply(500) 을 넣어준다.)
- 요청은 네트워크 요청이기 때문에 Dispatcher 종류를 적절하게 이용한다.
- 각 다운로드 작업은 "n번 파일 다운로드 시작" -> "n번 파일 다운로드 완료" 를 출력한다.
- 모든 파일이 전부 다운로드 완료 후에 총 걸린 시간을 출력한다.

````kotlin
@file:DependsOn("org.jetbrains.kotlinx:kotlinx-coroutines-core:1.6.4")
import kotlinx.coroutines.*
````

````kotlin
import kotlinx.coroutines.sync.*
import kotlin.time.Duration

suspend fun fetchData(url: String): Unit {
    println("$url 번 파일 다운로드 시작")
    delay(1000)
    println("$url 번 파일 다운로드 완료")
}

runBlocking {
    val start = System.currentTimeMillis()
    val semaphore = Semaphore(5)
    val jobs = (1..100).map { i ->
        async(Dispatchers.IO) {
            semaphore.withPermit {
                fetchData(i.toString())
            }
        }
    }
//    jobs.awaitAll()
    jobs.joinAll()
    val end = System.currentTimeMillis()
    println("총 걸린 시간: ${(end - start) / 1_000}s")
}
````

### 4장 과제 - 2
- 대규모 배치 작업 비동기 처리 시스템 시뮬레이션

요구사항
- 데이터 베이스에 5만건의 데이터가 있다고 한다. (1~50,000 번호로 대체)
- 유저 데이터를 가져오는데는 10ms 가 걸린다 가정한다.
- 시스템 과부하를 막기 위해 한번에 최대 100개의 작업만 가능
- 그리고 10% 확률로 실패할 수 있음 (랜덤 함수로 실패 구현)
 (Hint. Random.nextInt(100) < 10 를 사용한다)
- 실패한 데이터는 따로 저장한 뒤에 retry 처리할 수 있게 한다 (여기선 map이나 list로 저장)
- 마지막에 성공한 수, 실패한 수, retry 목록 을 출력한다.


#### 프로그램 결과 예시

````kotlin
import kotlinx.coroutines.sync.*
import kotlin.random.Random

val dataLength = 50_000
suspend fun fetchData(id: Int): Pair<Boolean, Int> {
    delay(20)
    val isFailed = Random.nextInt(100) < 10
    val samplingCount = 20
    val skipLoggingRange = 0 + samplingCount..dataLength - samplingCount
    if (id !in skipLoggingRange) {
        println("$id 번 유저 데이터 처리 ${if (isFailed) "실패" else "성공"}")
    }
    return Pair(isFailed, id)
}

suspend fun retryData(id: Int): Unit {
    delay(20)
//    println("재시도 처리: $id")
}

runBlocking {
    val semaphore = Semaphore(100)
    val failedIds = mutableListOf<Int>()
    val chunkSize = 1_000
    val start = System.currentTimeMillis()
    var successCounter = 0
    (1..dataLength).chunked(chunkSize).map { chunk ->
        val jobs = chunk.map { id ->
            async(Dispatchers.IO) {
                semaphore.withPermit {
                    fetchData(id)
                }
            }
        }
        jobs.awaitAll()
            .forEach {
                if (it.first) {
                    failedIds.add(it.second)
                } else {
                    successCounter++
                }
            }
    }

    failedIds.chunked(chunkSize).map { chunk ->
        val retryJob = chunk.map { id ->
            async(Dispatchers.IO) {
                semaphore.withPermit {
                    retryData(id)
                }
            }
        }
        retryJob.joinAll()
    }


    println("성공한 개수: $successCounter 개")
    println("실패한 개수: ${failedIds.size} 개")
    println("retry 목록: $failedIds")
    println("총 걸린 시간: ${(System.currentTimeMillis() - start) / 1_000}s")
}
````
